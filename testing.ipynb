```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

```

```python
my_data = pd.read_excel("Position_Salaries.xlsx")
my_data.head()
```

```python
x = my_data.iloc[:, -2:-1].values
y = my_data.iloc[:, -1].values
print(x.shape, y.shape)
y
```

```python
linear_regr = LinearRegression()
linear_regr.fit(x, y)
```

```python
y_pred_lin = linear_regr.predict(x)
```

```python
plt.scatter(x, y, marker="x", color="red")
plt.plot(x, y_pred_lin, color="blue")
plt.title("Linear Test")
plt.show()
```

```python
poly_reg = PolynomialFeatures(degree=4)
x_poly = poly_reg.fit_transform(x)
x_poly
```

```python
lin_reg_poly = LinearRegression()
lin_reg_poly.fit(x_poly, y)
y_pred_poly = lin_reg_poly.predict(x_poly)
```

```python
plt.scatter(x, y, marker="x", color="red")
plt.plot(x, y_pred_poly, color="blue")
plt.title("Polynomial Test")
plt.show()
```
